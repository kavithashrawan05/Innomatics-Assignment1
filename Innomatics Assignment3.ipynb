# Problem 1 : Employee Performance Bonus Eligibility


```python
# Given dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest score
highest_score = max(employees.values())

# Step 2: Find all employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

# Problem 2 : Search Query Keyword Analysis


```python
import string

# Given input
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency of each word
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Step 5: Filter words that appear more than once
result = {word: count for word, count in word_count.items() if count > 1}

# Display output
print(result)
```

# Problem 3 :  Sensor Data Validation


```python
# Given input
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings
valid_readings = []

# Loop through the list using enumerate to get index and value
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Check if even
        valid_readings.append((index, value))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem 4 : Email Domain Usage Analysis


```python
# Given input
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Step 1: Extract domain and count
for email in emails:
    domain = email.split("@")[1]   # Extract domain part
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate total emails
total_emails = len(emails)

# Step 3: Calculate percentage usage
print("Email Domain Usage Percentage:")
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

# Problem 5 : Sales Spike Detection

```python
# Given input
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Calculate 30% above average
threshold = average_sales * 1.30

# Step 3: Detect sales spikes
print("Sales Spike Days:")
for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")
```

# Problem 6 : Duplicate User ID Detection

```python
# Given input
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency of each user ID
id_count = {}

# Step 1: Count occurrences
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Display duplicates
print("Duplicate User IDs:")
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
