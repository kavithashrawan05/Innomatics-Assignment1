```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("❌ Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("❌ Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False
    else:
        print("✅ Recharge Successful!")
        return True


while True:
    try:
        recharge_amount = int(input("Enter recharge amount: ₹"))
        
        if validate_recharge(recharge_amount):
            break
        else:
            print("Please try again.\n")
            
    except ValueError:
        print("❌ Please enter a valid numeric amount.\n")
```

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product} → Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product} → Stock OK (Stock: {quantity})")


# Dictionary of products and stock quantities
inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 18,
    "Salt": 5,
    "Soap": 30
}

check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    # Determine result
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Example student marks
student_marks = [60, 45, 70, 55, 40]

avg, result = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)



```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)
    
    return fare


while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")
        
        total_fare = calculate_fare(distance, peak)
        
        print("Total Cab Fare: ₹", total_fare)
        
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break
            
    except ValueError:
        print("❌ Please enter valid numeric distance.\n")
```

```python
def check_eligibility(attendance_list):
    present_days = 0
    
    # Count present days
    for status in attendance_list:
        if status == "P":
            present_days += 1
    
    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Example attendance record
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

percent, result = check_eligibility(attendance)

print("Attendance Percentage:", percent)
print("Eligibility Status:", result)
```

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    
    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum length should be 8)"
    
    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    
    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


# Example password
pwd = input("Enter password: ")

result = check_password_strength(pwd)
print(result)
```
