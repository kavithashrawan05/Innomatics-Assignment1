# Task2

```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print(".....Result Analysis.....")
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

# Task3

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

# Task4

```python
messages = ["Hi", "Welcome to the platform", "OK"]
print("-----Message Length Analysis-----")
for message in messages:
    length = len(message)
    print(f"Message: '{message}'")
    print("Length:", length)
    if length>10:
        print("Status: Long Message")
    else:
        print("Status: Short Message")
    print()
```
