```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        if parked > peak_usage:
            peak_usage = parked

    # Check capacity
    if parked > capacity:
        status = "Over Capacity Alert!"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)


# Example
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
def delivery_time_estimator(distance, traffic, weather):
    # Base time = 5 minutes per km
    eta = distance * 5

    # Traffic adjustment
    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 8

    # Weather adjustment
    if weather == "Rainy":
        eta += 10
    elif weather == "Storm":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")


# Example
delivery_time_estimator(8, "High", "Rainy")
```

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = 0

    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", occupancy, "%")
    print("Show Status:", status)


# Example (Assume 150 seats booked)
theatre_occupancy(200, [1]*150)
```

```python
def server_load_classifier(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", average, "%")
    print("Server Status:", status)


# Example
server_load_classifier([45, 60, 70, 85, 90])
```

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:   # Threshold for overuse
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


# Example
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage_data)
```

```python
def event_registration_controller(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    if waitlisted > 0:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Example
event_registration_controller(100, 105)
```
