```python
# Function to analyze post engagement
def analyze_engagement(likes_list):
    
    total_likes = 0
    
    # Using loop to calculate total likes
    for like in likes_list:
        total_likes += like
    
    print("Total Likes:", total_likes)
    
    # Checking post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example Data (likes from different sources or days)
likes = [200, 150, 300, 400]  

analyze_engagement(likes)
```

```python
# Function to analyze rainfall data
def rainfall_checker(rainfall_data, required_level):
    
    total_rainfall = 0
    
    # Using loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain
    
    # Calculating average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)
    
    print("Average Rainfall:", int(average_rainfall))
    
    # Checking rainfall adequacy
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example Data (seasonal rainfall values)
rainfall_values = [60, 75, 80, 73]

# Required rainfall level
required = 70

rainfall_checker(rainfall_values, required)
```

```python
# Function to detect duplicate usernames
def detect_duplicates(usernames):
    
    unique_users = set(usernames)
    
    # If length differs, duplicates exist
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example Data
user_list = ["kavi123", "ram456", "anu789", "kavi123"]

detect_duplicates(user_list)
```

```python
# Function to filter premium crop prices
def filter_premium_crops(prices):
    
    premium_crops = []
    
    # Using list filtering logic
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)


# Example Data
crop_prices = [1500, 2500, 1800, 3200]

filter_premium_crops(crop_prices)
```

```python
# Function to evaluate application health
def check_system_health(error_count):
    
    print("Error Count:", error_count)
    
    # Conditional statements to check system status
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example usage
check_system_health(7)
```

```python
# Function to validate daily transaction amount
def check_transaction(amount):
    
    daily_limit = 50000
    
    print("Transaction Amount:", amount)
    
    # Conditional logic to check transaction status
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example usage
check_transaction(60000)
```

```python
# Function to calculate attendance percentage and check eligibility
def check_attendance(attendance_list):
    
    total_classes = len(attendance_list)
    attended_classes = 0
    
    # Using loop to count attended classes
    for status in attendance_list:
        if status == 1:   # 1 means present, 0 means absent
            attended_classes += 1
    
    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    
    print("Attendance Percentage:", attendance_percentage)
    
    # Check eligibility
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example Data (1 = Present, 0 = Absent)
attendance_data = [1, 1, 1, 1, 0]   # 4 out of 5 classes attended

check_attendance(attendance_data)
```

```python
# Function to calculate electricity bill and usage classification
def calculate_bill(units):
    
    bill = 0
    
    # Slab calculation
    if units <= 100:
        bill = units * 3
        
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
        
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status


# Example usage
units_consumed = 250
total_bill, usage_status = calculate_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Status:", usage_status)
```
