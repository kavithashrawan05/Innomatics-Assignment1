```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display(self):
        print(f"{self.light_name} is {self.status}")


# Sample Usage
light = SmartLight("Bedroom Light")
light.turn_on()
light.display()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Usage
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Usage
c1 = Contact("Anita", "9876543210")
c1.display()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample Usage
p1 = Product("Headphones", 2499)
p1.display_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample Usage
m1 = Movie("Inception", 4.8)
m1.display()
```
