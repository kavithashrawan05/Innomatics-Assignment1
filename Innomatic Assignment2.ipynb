# problem2

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

# problem3

```python
numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
        
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

# problem5

```python
# Given attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count total days
total_days = len(attendance)

# Count present days
present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

# problem7

```python
text = "python"
frequency = {}
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
print(frequency)
```
